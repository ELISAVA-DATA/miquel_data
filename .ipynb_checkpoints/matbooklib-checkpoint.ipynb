```python
import matplotlib.pyplot as plt
import random

meses = ["Enero", "Febrero", "Marzo", "Abril", "Mayo", "Junio", 
         "Julio", "Agosto", "Septiembre", "Octubre", "Noviembre", "Diciembre"]

valores = [random.randint(10, 100) for _ in meses]

colores = [f"C{i}" for i in range(len(meses))]  # Usar colores predefinidos de Matplotlib

# Crear gráfico de barras
plt.figure(figsize=(10, 6))
plt.bar(meses, valores, color=colores)
plt.xlabel("Meses del Año")
plt.ylabel("Valores")
plt.title("Meses del Año vs Valores")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

```
